```python
# Data Cleaning
```

```python
df.isnull().sum().reset_index()
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
```

```python
marks_df=df[["10th Mark","12th Mark","college mark"]]
marks_df.head()
```

```python
df["Average Marks"]=marks_df.mean(axis=1)
df[["Average Marks"]].head()
```

```python
high_perform=df[df["college mark"]>70]
high_perform.head()
```

```python
sorted_data=df.sort_values(by="college mark",ascending=False)
sorted_data.head()
```

```python
gender_avg=df.groupby("Gender")["college mark"].mean()
gender_avg.head().reset_index()
```

```python
correlation = df[["10th Mark", "12th Mark", "college mark"]].corr()
correlation.reset_index()
```

```python
def result_fun(x):
    if x >= 40:
        return "Pass"
    else:
        return "Fail"

df["Result"]=df["college mark"].map(result_fun)
df[["college mark","Result"]].head()
```
